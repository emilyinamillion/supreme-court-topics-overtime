```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
caseqtypost = pd.read_csv("data/post1946votes.csv", encoding = 'iso-8859-1')
caseqtypre = pd.read_csv("data/pre1946votes.csv", encoding = 'iso-8859-1')
casefull = pd.concat([caseqtypost, caseqtypre] )
```

```python
casefull
```

```python
def extractyear(text):
    text = text.split("-")
    return int(text[0])

casefull["year"] = casefull.caseId.apply(extractyear)
casefull.sort_values("year", ascending = True, inplace = True)
casefull
```

```python
foranalysis = casefull.groupby("year").count()
```

```python
#foranalysis = foranalysis[["caseId"]]
foranalysis.reset_index(inplace = True)
```

```python
import re
def cleanup(text):
    if text < 1800:
        group = 1790
    if text >= 1800 and text < 1810:
        group = 1800
    if text >= 1810 and text < 1820:
        group = 1810
    if text >= 1820 and text < 1830:
        group = 1820
    if text >= 1830 and text < 1840:
        group = 1830
    if text >= 1840 and text < 1850:
        group = 1840
    if text >= 1850 and text < 1860:
        group = 1850
    if text >= 1860 and text < 1870:
        group = 1860
    if text >= 1870 and text < 1880:
        group = 1870
    if text >= 1880 and text < 1890:
        group = 1880
    if text >= 1890 and text < 1900:
        group = 1890
    if text >= 1900 and text < 1910:
        group = 1900
    if text >= 1910 and text < 1920:
        group = 1910
    if text >= 1920 and text < 1930:
        group = 1920
    if text >= 1930 and text < 1940:
        group = 1930
    if text >= 1940 and text < 1950:
        group = 1940
    if text >= 1950 and text < 1960:
        group = 1950
    if text >= 1960 and text < 1970:
        group = 1960
    if text >= 1970 and text < 1980:
        group = 1970
    if text >= 1980 and text < 1990:
        group = 1980
    if text >= 1990 and text < 2000:
        group = 1990
    if text >= 2000 and text < 2010:
        group = 2000
    if text >= 2010 and text < 2015:
        group = 2010
    return group

foranalysis["decade"] = foranalysis.year.apply(cleanup)
foranalysis
```

```python
new = foranalysis.groupby("decade").sum()
new.reset_index(inplace = True)
```

```python
new.to_json("justdecades.json")
```
