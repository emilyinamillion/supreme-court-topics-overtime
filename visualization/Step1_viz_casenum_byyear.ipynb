## For the visualization, we need the total cases every year over the course of the court

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#import seaborn as sns
```

```python
caseqtypost = pd.read_csv("data/post1946votes.csv", encoding = 'iso-8859-1')
caseqtypre = pd.read_csv("data/pre1946votes.csv", encoding = 'iso-8859-1')
casefull = pd.concat([caseqtypost, caseqtypre] )
```

```python
casefull
```

```python
def extractyear(text):
    text = text.split("-")
    return int(text[0])

casefull["year"] = casefull.caseId.apply(extractyear)
casefull.sort_values("year", ascending = True, inplace = True)
casefull
```

```python
foranalysis = casefull.groupby("year").count()
```
